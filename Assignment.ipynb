```python
state_cases = pd.read_csv('state_level_latest.csv')
state_cases
```

```python
state_warning = pd.read_csv('HospitalBedsIndia.csv')
state_warning
```

```python
state_warning.columns
```

```python

BGujarat = state_warning['NumPublicBeds_HMIS'].iloc[11]
BMaharashtra = state_warning['NumPublicBeds_HMIS'].iloc[20]
BKarnataka = state_warning['NumPublicBeds_HMIS'].iloc[16]
BWestBengal = state_warning['NumPublicBeds_HMIS'].iloc[35]
BKerala = state_warning['NumPublicBeds_HMIS'].iloc[17]
BDelhi = state_warning['NumPublicBeds_HMIS'].iloc[9]
BAssam = state_warning['NumPublicBeds_HMIS'].iloc[3]
BBihar = state_warning['NumPublicBeds_HMIS'].iloc[4]
BUttarakhand = state_warning['NumPublicBeds_HMIS'].iloc[34]
BTelangana = state_warning['NumPublicBeds_HMIS'].iloc[31]
BMadhyaPradesh = state_warning['NumPublicBeds_HMIS'].iloc[19]
BRajasthan = state_warning['NumPublicBeds_HMIS'].iloc[28]
BPunjab = state_warning['NumPublicBeds_HMIS'].iloc[27]
BUttarPradesh = state_warning['NumPublicBeds_HMIS'].iloc[33]
 
beds = [BGujarat,BMaharashtra,BKarnataka,BWestBengal,BKerala,BDelhi,BAssam,BBihar,BUttarakhand,BTelangana,BMadhyaPradesh,BRajasthan,BPunjab,BUttarPradesh]
```

```python
covid = pd.DataFrame({'Beds':beds})
covid
```

```python
state_cases.columns
```

```python
states = state_cases.sort_values(by=['Confirmed'], ascending = False)
states
```

```python

CGujarat = states['Confirmed'].iloc[8]
CMaharashtra = states['Confirmed'].iloc[1]
CKarnataka = states['Confirmed'].iloc[5]
CWestBengal = states['Confirmed'].iloc[7]
CKerala = states['Confirmed'].iloc[16]
CDelhi = states['Confirmed'].iloc[3]
CAssam = states['Confirmed'].iloc[12]
CBihar = states['Confirmed'].iloc[10]
CUttarakhand = states['Confirmed'].iloc[21]
CTelangana = states['Confirmed'].iloc[9]
CMadhyaPradesh = states['Confirmed'].iloc[14]
CRajasthan = states['Confirmed'].iloc[11]
CPunjab = states['Confirmed'].iloc[18]
CUttarPradesh = states['Confirmed'].iloc[6]

confirmed = [CGujarat,CMaharashtra,CKarnataka,CWestBengal,CKerala,CDelhi,CAssam,CBihar,CUttarakhand,CTelangana,CMadhyaPradesh,CRajasthan,CPunjab,CUttarPradesh]
```

```python
covid['Confirmed'] = confirmed
covid
```

```python

RGujarat = states['Recovered'].iloc[8]
RMaharashtra = states['Recovered'].iloc[1]
RKarnataka = states['Recovered'].iloc[5]
RWestBengal = states['Recovered'].iloc[7]
RKerala = states['Recovered'].iloc[16]
RDelhi = states['Recovered'].iloc[3]
RAssam = states['Recovered'].iloc[12]
RBihar = states['Recovered'].iloc[10]
RUttarakhand = states['Recovered'].iloc[21]
RTelangana = states['Recovered'].iloc[9]
RMadhyaPradesh = states['Recovered'].iloc[14]
RRajasthan = states['Recovered'].iloc[11]
RPunjab = states['Recovered'].iloc[18]
RUttarPradesh = states['Recovered'].iloc[6]

Recovered = [RGujarat,RMaharashtra,RKarnataka,RWestBengal,RKerala,RDelhi,RAssam,RBihar,RUttarakhand,RTelangana,RMadhyaPradesh,RRajasthan,RPunjab,RUttarPradesh]
```

```python
covid['Recovered'] = Recovered
covid
```

```python
covid.index = ['Gujarat','Maharashtra','Karnataka','WestBengal','Kerala','Delhi','Assam','Bihar','Uttarakhand','Telangana','Madhya Pradesh','Rajasthan','Punjab','Uttar Pradesh']
covid
```

```python
plt.figure(figsize = (9,14))
x = covid.index.tolist()
y1 = covid['Beds'].tolist()
y2 = covid['Confirmed'].tolist()
y4 = covid['Recovered'].tolist()

plt.plot(x, y1, '-s', label = "Number of Beds available",color = "blue")
plt.xticks(rotation = 90)
plt.plot(x, y2, '-o', label = "Number of Confirmed Cases of Covid-19",color = "red")
plt.plot(x, y4, '-^', label = "Number of Recovered Cases",color = "green")

plt.title('Overall Beds and Cases of Covid-19')
 
plt.legend(frameon=False)
[plt.gca().spines[loc].set_visible(False) for loc in ['top', 'right']]
#sns.color_palette("RdBu", n_colors=7)
```
